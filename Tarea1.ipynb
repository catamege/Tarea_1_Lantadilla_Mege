### Pregunta 1.5

Para determinar si las variables son "útiles" y sus valores son "correctos" es necesario comprender cada uno de los atributos del dataset.
Investigue y explique brevemente la relación **teórica** entre el `GDP (% per capita)` y cada una de las variables denominadas como "Características del país" en la introducción.

Ejemplo: Existe una variable denominada `Coastline (coast/area ratio)`. Coastline es una medida de la cantidad de costa (acceso a mar) del país normalizada al área total del país para no beneficiar a países más grandes pero con la misma proporción de costa. A mayor "Costline" aumenta la capacidad portuaria per capita del país, más puertos facilita el comercio y podría aumentar el GDP per cápita.

---


**Area (sq. mi.) :** Según la evidencia disponible, no existe una relación clara entre el tamaño de un país y su PIB per cápita. Aunque podría pensarse que un país más grande tendría más recursos disponibles, lo que realmente afecta es la calidad de las instituciones y la eficiencia en el uso de esos recursos. Esto se puede explicar en papers, tales como "Country size and economic growth and development" de Nadim G. Khalaf

**Pop. Density (per sq. mi.):** Según "How does population growth affect economic growth and vice versa? An empirical analysis" por Juan Gabriel Brida, Emiliano Alvarez, Gaston Cayssials y Matias Mednik publicado el año 2024, el tamaño de la población tiene una fuerte relación con el PIB y el PIB per cápita, sin embargo la interdependencia entre el cambio demográfico y el crecimiento económico no se puede explicar con un único modelo y, en cambio, depende de cada país. También sugiere que esta relación depende de otros factores, como capital humano e innovación del país.

**Coastline (coast/area ratio):** Tener acceso a la costa puede influir positivamente en el crecimiento económico, pues facilita el comercio y el acceso a recursos marinos. En efecto, cerca de todos los países sin borde costero (excluyendo a Europa) tienen lentos crecimientos económicos en comparación a países que sí tienen costa y que son parecidos en otras características. Esta relación la explora el autor Aashiqullah Zazai en "Economic Growth of Landlocked Countries" el año 2020. Por otro lado, entre países que sí tienen acceso al mar, el tamaño de la costa no tiene una relación significativa con el PIB, en cambio, importan otros factores como la infrastructura y la gestión costera, los autores Davide Pasquali y Alessandro Marucci investigan esta idea en el paper "The Effects of Urban and Economic Development on Coastal Zone Management" publicado el año 2021.

**Net migration:** De manera general, la migración tiene pocos efectos en el PIB per cápita, tal como lo explica Frederick Treyz y Peter Evangelakis en "Immigration and United States Economic Growth". No obstante, publicado en el National Bereau of Economics, el autor George Borjas explica que la migración neta puede tener un efecto positivo en el crecimiento si es que está compuesto de una entrada de trabajadores altamente calificados, es decir importa la composición de habilidades de los migrantes y la distribución de años en el mercado laboral. 

**Infant mortality (per 1000 births):** La mortalidad infantil tiene una relación asimétrica con el PIB, pues el crecimiento económico positivo puede tener efectos débiles y mixtos en la reducción de la mortalidad infantil, pero el crecimiento económico negativo tiene un fuerte impacto adverso. Esto lo profundiza Akira Nishiyama en "Economic Growth and Infant Mortality in Developing Countries".

**Literacy (%):** El alfabetismo se asocia a capital humano y capital humano a productividad, por lo que mayor una mayor tasa de alfabetización generalmente se asocia con un mayor PIB per cápita. Esto lo discuten Helena Weatherly, Kendy Lopez y Carol Tierra en "The Impact of Education on GDP per Capita".

**Phones (per 1000):** Cantidad de telefonos cada 100 personas, puede estar asociado al nivel de avance tecnológico del país. De esta manera, el aumento en la penetración de teléfonos móviles está positivamente correlacionado con el crecimiento del PIB per cápita. Esto lo investiga Ghirmai Kefela en "The impact of mobile phone and economic growth in developing countries "

**Arable (%):** El porcentaje de tierra adecuada para la agricultura tiene una relación directa con capacidad del país para irse a este tipo de industria. Bajo esta idea, la agricultura tiene un impacto positivo en el PIB per cápita, Gjergji Mero1, Remzi Keco, Myslym Osmani y Arben Kambo explican sobre la importancia de la agricultura en "Role of Agriculture to GDP and National per Capita Income Growth-What do Global Data Reveal for Albania? "

**Crops (%):** El porcentaje de los cultivos de un país también tiene una relación con la agricultura y así se puede concluir que hay una relación positiva entre los cultivos y el PIB.

**Other (%):**

**Climate:** Toma valores 1, 2, 3 y 4. Cada uno probablemente represente un tipo de clima como tropial, templado, desértico, etc. Un clima favorable para la producción, puede implicar mayores niveles de PIB per cápita.

**Birthrate:** Representa la tasa de natalidad, según la literatura económica, en general aquellos países con mayor PIB per cápita, tienen menores tasas de natalidad, probablemente porque tienen menos tasas de embarazo adolescente y mayor educación sexual.

**Deathrate:** Representa la tasa de mortailidad, según la literatura económica, en general aquellos países con mayor PIB per cápita, tienen menores tasas de mortalidad porque probablemente tengan mejores sitemas de salud y sanidad.

**Agriculture, Industry y Service**: Son variables que toman valores entre 0 y 1, probablemente representen el % del ingreso del país que proviene del sector agrícola, industrial y de servicios, respectivamente.
