**Area (sq. mi.) :** Según la evidencia disponible, no existe una relación clara entre el tamaño de un país y su PIB per cápita. Aunque podría pensarse que un país más grande tendría más recursos disponibles, lo que realmente afecta es la calidad de las instituciones y la eficiencia en el uso de esos recursos. Esto se puede explicar en papers, tales como "Country size and economic growth and development" de Nadim G. Khalaf

**Pop. Density (per sq. mi.):** Según "How does population growth affect economic growth and vice versa? An empirical analysis" por Juan Gabriel Brida, Emiliano Alvarez, Gaston Cayssials y Matias Mednik publicado el año 2024, el tamaño de la población tiene una fuerte relación con el PIB y el PIB per cápita, sin embargo la interdependencia entre el cambio demográfico y el crecimiento económico no se puede explicar con un único modelo y, en cambio, depende de cada país. También sugiere que esta relación depende de otros factores, como capital humano e innovación del país.

**Coastline (coast/area ratio):** Tener acceso a la costa puede influir positivamente en el crecimiento económico, pues facilita el comercio y el acceso a recursos marinos. En efecto, cerca de todos los países sin borde costero (excluyendo a Europa) tienen lentos crecimientos económicos en comparación a países que sí tienen costa y que son parecidos en otras características. Esta relación la explora el autor Aashiqullah Zazai en "Economic Growth of Landlocked Countries" el año 2020. Por otro lado, entre países que sí tienen acceso al mar, el tamaño de la costa no tiene una relación significativa con el PIB per cápita, en cambio, importan otros factores como la infrastructura y la gestión costera, los autores Davide Pasquali y Alessandro Marucci investigan esta idea en el paper "The Effects of Urban and Economic Development on Coastal Zone Management" publicado el año 2021.

**Net migration:** De manera general, la migración tiene pocos efectos en el PIB per cápita, tal como lo explica Frederick Treyz y Peter Evangelakis en "Immigration and United States Economic Growth". No obstante, publicado en el National Bereau of Economics, el autor George Borjas explica que la migración neta puede tener un efecto positivo en el crecimiento si es que está compuesto de una entrada de trabajadores altamente calificados, es decir importa la composición de habilidades de los migrantes y la distribución de años en el mercado laboral. 

**Infant mortality (per 1000 births):** La mortalidad infantil tiene una relación asimétrica con el PIB per cápita, pues el crecimiento económico positivo puede tener efectos débiles y mixtos en la reducción de la mortalidad infantil, pero el crecimiento económico negativo tiene un fuerte impacto adverso. Esto lo profundiza Akira Nishiyama en "Economic Growth and Infant Mortality in Developing Countries".

**Literacy (%):** El alfabetismo se asocia a capital humano y capital humano a productividad, por lo que mayor una mayor tasa de alfabetización generalmente se asocia con un mayor PIB per cápita. Esto lo discuten Helena Weatherly, Kendy Lopez y Carol Tierra en "The Impact of Education on GDP per Capita".

**Phones (per 1000):** Cantidad de telefonos cada 1000 personas, puede estar asociado al nivel de avance tecnológico del país. De esta manera, el aumento en la poseción de teléfonos móviles está positivamente correlacionado con el crecimiento del PIB per cápita. Esto lo investiga Ghirmai Kefela en "The impact of mobile phone and economic growth in developing countries "

**Arable (%):** El porcentaje de tierra adecuada para la agricultura tiene una relación directa con capacidad del país para irse a este tipo de industria. Bajo esta idea, la agricultura tiene un impacto positivo en el PIB per cápita, Gjergji Mero1, Remzi Keco, Myslym Osmani y Arben Kambo explican sobre la importancia de la agricultura en "Role of Agriculture to GDP and National per Capita Income Growth-What do Global Data Reveal for Albania? "

**Crops (%):** El porcentaje de los cultivos de un país también tiene una relación con la agricultura y así se puede concluir que hay una relación positiva entre los cultivos y el PIB per cápita, tal como se explicó en el inciso anterior.

**Other (%):** Suponiendo que el porcentaje restante es el porcentaje de urbanización, habrá una relación positiva de esta variable con el PIB per cápita. Los autores Mingxing Chen, Hua Zhang, Weidong Liu y Wenzhong Zhang explican en detalle esta relación con el paper "The Global Pattern of Urbanization and Economic Growth: Evidence from the Last Three Decades", aquí mencionan que la mayoría de países con altos PIB per cápita son altamente urbanizados, sin embargo, la velocidad de urbanización no siempre se traduce en un crecimiento económico acelerado, especialmente si esta es impulsada por el gobierno como motor principal.

**Climate:** Toma valores 1, 2, 3 y 4. Cada uno probablemente represente un tipo de clima como tropial, templado, desértico, etc. Un clima favorable para la producción, puede implicar mayores niveles de PIB per cápita. En específico, el estudio "Climate, Water and Agriculture: Impacts on and Adaptation of Agro-Ecological Systems in the Tropics" por Robert Mendelsohn y Ariel Dinar explica que el tipo de clima de un país puede influir significativamente en la productividad agrícola, y otros sectores económicos, provocando, en consencuencia, que tenga una relación positiva con el PIB.

**Birthrate:** Representa la tasa de natalidad, según la literatura económica, en general aquellos países con mayor PIB per cápita, tienen menores tasas de natalidad, probablemente porque tienen menos tasas de embarazo adolescente y mayor educación sexual. Un ejemplo de esta literatura es "Influence of GDP per Capita on Birth Rates" por Derek Richard Koppes, que explora la relación entre la tasa de natalidad y el PIB per cápita en 108 países, encontrando que los países con mayor PIB per cápita tienen mayor disponibilidad de recursos de planificación familiar, por lo que la tasa de natalidad de estos países es menor.

**Deathrate:** Representa la tasa de mortailidad, según la literatura económica, en general aquellos países con mayor PIB per cápita, tienen menores tasas de mortalidad porque probablemente tengan mejores sitemas de salud, alcantarillado y sanidad en general. Un ejemplo de esta literatura es el paper "Effect of GDP Per Capita on National Life Expectancy" por Varsha Gokare, donde explica que países más ricos tienen menor cantidad de muertes, tanto en niños como en adultos.

**Agriculture, Industry y Service**: Son variables que toman valores entre 0 y 1, probablemente representen el % del ingreso del país que proviene del sector agrícola, industrial y de servicios, respectivamente. Como se explico anteriormente, la agricultura tiene una relación positiva con el PIB y el PIB per cápita. Por otro lado, el nivel de industralización también tiene una relación positiva con las variables anteriores, pues como lo menciona Dongchul Cho en "Industrialization, Convergence, and Patterns of Growth", publicado el año 1994 en el Southern Economic Journal, la industrialización puede impulsar el crecimiento económico y mejorar así la productividad y el PIB per cápita. Finalmente el porcentaje de servicios dentro de un país generalmente se asocia con un mayor PIB per cápita, esto lo investigan más en profundidad en el paper "The Two Waves of Service Sector Growth" por Barry Eichengreen y Poonam Gupta, donde identifican el crecimiento del PIB per cápita producto por dos olas de crecimiento del sector servicios: una en países con niveles relativamente bajos de PIB per cápita y otra en países con ingresos per cápita más altos, el estudio concluye que existe una relación más fuerte en el impulso del sector de servicios y el PIB en los países más pobres.

### Pregunta 1.8

El df contiene variables con missing values (`NaN`). Impute los `NaN` con el método que estime conveniente, justificando su decisión.

¿Es pertinente eliminar alguna de estas variables? Hágalo si es el caso.

```python
#Identificamos las variables con missing
df.isna().sum()

# Con 227 filas no hay ninguna variable (columna) en particular que debería 
# eliminarse, ya que el % de missings por variable es menor al 10%. Además, 
# consideramos que son variables que contienen información relevante para el 
# análisis. 
```

```python
#Identificamos cuáles son los países tienen missings en cada variable 

for col in df.columns:
    if df[col].isnull().sum() > 0:
        print(f"Países con missing en la variable: {col}")
        print(df[df[col].isnull()]["Country"])
```

```python
# Contar el número de valores faltantes para cada país
missing_counts = df.isnull().sum(axis=1)

# Identificar países con más del 25% de variables faltantes
threshold = len(df.columns) / 4
countries_to_remove = df[missing_counts >= threshold]

countries_to_remove #son 3 de 227
```

```python
# Eliminar países con más del 25% de variables faltantes
df = df[missing_counts < threshold].reset_index(drop=True)
```

---

Consideramos aceptable eliminar los países con más de un 25% de variables sin datos. En primer lugar porque son 3 países de distintas regiones, lo que es un 1.3% de la muestra, por lo que su eliminación no va a afectar significativamente la representatividad de la muestra. En segundo lugar, porque al ser países con muchas variables relevantes sin datos, su imputación podría generar sesgos en futuras estimaciones.

---

```python
#Aún quedan missings, veremos si los paises que tienen missings son de una 
#region en especifico o no, para asi saber si funciona imputar una mediana 
#por region a los valores missings.

missing_by_region = df.groupby("Region").apply(lambda x: x.isnull().sum())

# Mostrar los valores faltantes por región
missing_by_region

#Western Europe es la region con mayor cantidad de missings pero aun asi
# son pocos y se distribuyen en distintas variables (lo máximo es 7 missings 
# de 21 países de WE para la variable climate), es decir se puede utilizar la 
# mediana de la región para imputar los missings ya que hay suficientes datos 
# por región para cada variable con missings. 
```

```python
#Imputamos con la mediana de cada region, porque asi nos olvidamos de que
# salte la media por valores extremos (vimos uno por uno los histogramas, pero
# dejamos GDP y Literacy para que lo demuestren)

# Las variables que no tienen valores extremos (Net migration, Climate,
# Deathrate,Industry y Service) tienen mediana y media parecida asi que esta ok
columns_to_impute = [
   "Net migration", "Infant mortality (per 1000 births)", "GDP ($ per capita)",
    "Literacy (%)", "Phones (per 1000)", "Arable (%)", "Crops (%)",
    "Other (%)", "Climate", "Birthrate", "Deathrate",
    "Agriculture", "Industry", "Service"
]

for column in columns_to_impute:
    df[column] = (
        df.groupby("Region")[column].transform(lambda x: x.fillna(x.median()))
    )
```

---

Decidimos imputar los missings restantes utilizando la mediana de la variable por región. En primer lugar, porque la media es sensible a los outliers (por ejemplo el observado en GDP) y puede no ser representativa del país con variables missing. Al imputar por la mediana, el valor se acerca más al comportamiento típico de la región. Además, este método permite mantener las distribuciones de las variables que no son simétricas, donde la media y mediana difieren. 

También decidimos imputar por la mediana de la región y no la mediana global porque de esta forma estamos capturando las diferencias que las regiones tienen en términos económicos, demográficos y sociales.

### Pregunta 1.9

¿Cómo distribuye el `GDP ($ per capita)` en diferentes **regiones**? Defina una forma ilustrativa de gráficar el `GDP ($ per capita)` para todas las regiones en un mismo gráfico. Interprételo.

```python
regiones = df["Region"].unique()
palette = sns.color_palette("hls", len(regiones))

plt.figure(figsize=(14, 7))

for i, region in enumerate(regiones):
    subset = df[df["Region"] == region]
    sns.kdeplot(
        subset["GDP ($ per capita)"],
        label=region,
        color=palette[i]
    )

# Escala logarítmica para mejor visualización
plt.xscale("log")
plt.title("Distribución de densidad del GDP per cápita por región (2007)")
plt.xlabel("GDP ($ per capita)[escala log]")
plt.ylabel("Densidad")
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

---

### Pregunta 2.6

Calcule el promedio a lo largo de toda la muestra ($\mathbb{E}[\cdot]$) para el cambio porcentual de cada país y genere una nueva serie con la resta entre el cambio porcentual del país $i$ en el año $t$, y el promedio del cambio porcentual del país $i$. En otras palabras, genere una serie con _**desvíos del cambio porcentual promedio**_ $\forall i,t$:

$$Nueva Serie_i = \Delta \% TotalCO2_{i,t} - \mathbb{E}[{\Delta \% TotalCO2_{i,t}}]$$

Luego, para los siguientes países:

- Reino Unido
- Canadá
- Alemania
- Francia
- Estados Unidos
- Japón
- Italia
- España


grafique en un panel _1x2_ la desviación del cambio porcentual respecto al promedio entre 2007 y 2010 en lado izquierdo, y entre 2017 y 2020 en el lado derecho (Hint: Hay comandos que facilitan esta labor. Puede intentar con `fig, axes = plt.subplots(1, 2, figsize=(18, 6), sharey=True)`, por ejemplo).

¿Existe algún país en particular que mostró mayores desviaciones atípicas de emisión de CO2 durante el periodo de la crisis sub-prime? ¿Cómo es el comportamiento de las desviaciones atípicas de CO2 de este país durante la crisis del Covid-19?

```python
# Calcular el promedio del cambio porcentual de CO2 por país
df_co2["mean_change"] = df_co2.groupby("Country")["%_change"].transform("mean")
```

```python
#desviación cambio porcentual
df_co2["deviation"] = df_co2["%_change"] - df_co2["mean_change"]
```

```python
# Filtrar datos
year_2007 = 2007
year_2010 = 2010
year_2017 = 2017
year_2020 = 2020
countries = ["United Kingdom", "Canada", "Germany", "France", "United States",
             "Japan", "Italy", "Spain"]
df_co2_2007_2010 = df_co2[(df_co2["date"].dt.year >= year_2007) &
                          (df_co2["date"].dt.year <= year_2010) &
                          (df_co2["Country"].isin(countries))]
df_co2_2017_2020 = df_co2[(df_co2["date"].dt.year >= year_2017) &
                          (df_co2["date"].dt.year <= year_2020) &
                          (df_co2["Country"].isin(countries))]

# Crear gráfico con dos paneles
fig, axes = plt.subplots(1, 2, figsize=(18, 6), sharey=True)

# Panel 1: 2007-2010
for country in countries:
    subset = df_co2_2007_2010[df_co2_2007_2010["Country"] == country]
    axes[0].plot(subset["date"], subset["deviation"], label=country)

axes[0].set_title("Desviación Cambio % CO2 (2007-2010)")
axes[0].set_xlabel("Año")
axes[0].set_ylabel("Desviación del Cambio % CO2")
axes[0].legend()

# Panel 2: 2017-2020
for country in countries:
    subset = df_co2_2017_2020[df_co2_2017_2020["Country"] == country]
    axes[1].plot(subset["date"], subset["deviation"], label=country)

axes[1].set_title("Desviación Cambio % CO2 (2017-2020)")
axes[1].set_xlabel("Año")
axes[1].legend()

plt.show()
```

---
En el gráfico de la izquierda, la línea verde representa a Alemania y muestra desviaciones atípicas significativas en sus emisiones de CO2 durante el período de la crisis subprime (reduce sus emisiones en aprox 300% comparado al promedio del cambio porcentual del país). Esto sugiere que Alemania experimentó fluctuaciones notables en sus emisiones de CO2, que pueden estar relacionadas con la disminución de la demanda global y las interrupciones en el comercio internacional durante la crisis.

Durante la crisis del 2020. Este país nuevamente tiene desviaciones atípicas, se puede ver en la línea verde que también reduce significativamente sus emisiones respecto al promedio del cambio porcentual del país.
